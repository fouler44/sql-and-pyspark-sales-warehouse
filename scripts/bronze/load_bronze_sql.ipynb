# Load bronze layer

The notebook uses `INSERT OVERWRITE` statements to load data into the bronze tables. This operation replaces any existing data in the target tables with the new data from the source, ensuring the bronze layer always reflects the latest available information.

## Loading crm tables

#### Insert data into `sales_project.bronze.crm_cust_info`

```python
import time

start_time = time.time()
```

```python
%sql
INSERT OVERWRITE 
  sales_project.bronze.crm_cust_info
SELECT
  cst_id,
  cst_key,
  cst_firstname,
  cst_lastname,
  cst_gndr,
  cst_marital_status,
  cst_create_date
FROM 
  sales_project.google_drive.cust_info;
```

#### Insert data into `sales_project.bronze.crm_prd_info`

```python
%sql
INSERT OVERWRITE 
  sales_project.bronze.crm_prd_info
SELECT
  prd_id,
  prd_key,
  prd_nm,
  prd_cost,
  prd_line,
  prd_start_dt,
  prd_end_dt
FROM 
  sales_project.google_drive.prd_info;
```

#### Insert data into `sales_project.bronze.crm_sales_details`

```python
%sql
INSERT OVERWRITE 
  sales_project.bronze.crm_sales_details
SELECT
  sls_ord_num,
  sls_prd_key,
  sls_cust_id,
  sls_order_dt,
  sls_ship_dt,
  sls_due_dt,
  sls_sales,
  sls_quantity,
  sls_price
FROM
  sales_project.google_drive.sales_details;
```
